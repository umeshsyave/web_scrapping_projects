```python
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_binary
import csv
```

```python
def my_url(keyword,page_no):
    keyword=keyword.replace(' ','+')
    url=f"https://www.flipkart.com/search?q={keyword}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={page_no}"
    return url

def get_data(obj):
    soups_data=obj.find_all('div',{'class':'_4ddWXP'})
    
    records=[]
    for item in soups_data:
        
        #Title_description.
        try:
            title_description=item.find('a',{'class':'s1Q9rs'}).string
        except:
            title_description=None
            
            
        #Flipkart Assured / Not Assured.
        if item.find('div',{'class':'_32g5_j'}):
            trust='Flipkart Assured'
        else:
            trust='Not Assured'
            
        #selling price.
        try:
            selling_price=item.find('div',{'class':'_30jeq3'}).string
        except:
            selling_price=None
            
            
        #original cost.
        try:
            original_cost=item.find('div',{'class':'_3I9_wc'}).string
        except:
            try:
                original_cost=item.find('div',{'class':'_30jeq3'}).string
            except:
                original_cost=None
        
        
        #discount.
        try:
            discount=item.find('div',{'class':'_3Ay6Sb'}).string.split()[0]
        except:
            discount='0%'
        
        
        #Rating.
        try:
            ratings=item.find('div',{'class':'_3LWZlK'}).text
        except:
            ratings=None
            
        
        #number of ratings
        try:
            total_ratings=item.find('span',{'class':'_2_R_DZ'}).text[1:-1]
        except:
            total_ratings=None
            
        records.append([title_description,trust,selling_price,original_cost,discount,ratings,total_ratings])
       
    return records

def results(record):
    with open('results.csv','w',newline='',encoding='utf-8') as file:
        writer=csv.writer(file)
        writer.writerow(['title_description','trust','selling_price','original_cost','discount','ratings','total_ratings'])
        writer.writerows(record)
        
    
def scrap_data(product,page):
    records=[]
    for num in range(1,int(page)):
        url=my_url(product,num)
        driver.get(url)
        soup=BeautifulSoup(driver.page_source,'html.parser')
        for item in get_data(soup):
            records.append(item)
    results(records)
```

```python
scrap_data('smart tracker',15)

```
